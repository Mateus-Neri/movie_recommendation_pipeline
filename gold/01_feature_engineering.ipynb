```python
import pandas as pd
```

```python
#Importando tabelas
df = spark.read.table('workspace.silver.recomendation_db_silver')
df = df.toPandas()
```

```python
genres = {
      28: "Action",
      12: "Abenteuer",
      16: "Animation",
      35: "Komödie",
      80: "Krimi",
      99: "Dokumentarfilm",
      18: "Drama",
      10751 :"Familie",
      14: "Fantasy",
      36: "Historie",
      27: "Horror",
      10402: "Musik",
      9648: "Mystery",
      10749: "Liebesfilm",
      878 : "Science Fiction",
      10770: "TV-Film",
      53: "Thriller",
      10752: "Kriegsfilm",
      37: "Western",
  }

df["genres_text"] = df["genre_ids"].apply(
    lambda ids: " ".join([genres.get(i, "") for i in ids])
)

df = df.drop('genre_ids', axis=1)
```

```python
#Criando a coluna 'content' onde ficarão todas informações de cada filme
df['content'] = df['overview'] +" "+ df['keywords']*2 +" "+ df['genres_text']*3
```

```python
display(df)
```

```python
df = df.drop(columns= ["keywords", "overview", "genres_text"])
```

```python
df = spark.createDataFrame(df)
df.write.mode('overwrite').option("mergeSchema", "true").saveAsTable('workspace.gold.recomendation_db_gold')
```
