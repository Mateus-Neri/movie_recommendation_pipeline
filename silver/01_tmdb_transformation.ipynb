```python
import pandas as pd
```

```python
#Importando tabela
df = spark.read.table('workspace.bronze.recomendation_db_bronze')
df = df.toPandas()
```

```python
#Limpeza
df = (df[df['overview'] != ""])
df = df.drop_duplicates(subset='id')
df = df[df["genre_ids"].apply(len) > 0]
```

```python
df = spark.createDataFrame(df)
df.write.mode('overwrite').option("mergeSchema", "true").saveAsTable('workspace.silver.recomendation_db_silver')
```
