```python
import pandas as pd
```

```python
#Importando tabelas
movies_info = spark.read.table('workspace.bronze.movies_info_bronze')
movies_info = movies_info.toPandas()
```

```python
#Pega somente o ano do lançamento
movies_info['realese_year'] = movies_info['realese_year'].str.slice(0, 4)
```

```python
movies_info_spark = spark.createDataFrame(movies_info)
movies_info_spark.write.mode('overwrite').option("mergeSchema", "true").saveAsTable('workspace.silver.movies_info_silver')
movies_info_spark.coalesce(1).write.mode("overwrite").parquet("/Volumes/workspace/artifacts/recommendation_artifacts/movies_info")
```
